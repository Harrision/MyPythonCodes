# Pythagorean Triplet

Given an array arr of N integers, write a function that returns true if there is a triplet (a, b, c) that satisfies a^2 + b^2 = c^2, otherwise false.

Example 1:

Input:
N = 5
Arr[] = {3, 2, 4, 6, 5}
Output: Yes
Explanation: a=3, b=4, and c=5 forms a
pythagorean triplet.

```python
n= int(input("N: "))
arr=[int(i) for i in input(f"Enter Array of {n} elements: ").split()]
```

```python
def Pythagorean(arr,n):
    square= []
    for i in range(len(arr)):
        square.append(arr[i]**2)
    
    print(square)
```

```python
Pythagorean(arr,n)
```
